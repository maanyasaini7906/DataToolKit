***DATA TOOL-KIT IN PYTHON:***

Q1.)  What is NumPy, and why is it widely used in Python?

Ans.) NumPy (Numerical Python) is a library for numerical computing in      
     Python. It provides support for arrays, matrices, and high-level mathematical functions. It’s fast and efficient, making it essential for scientific and machine learning applications.

Q2) How does broadcasting work in NumPy?

Ans.) Broadcasting allows NumPy to perform operations on arrays of    
     different shapes by stretching the smaller array to match the larger one without making actual copies. This enables element-wise operations without looping manually.

Q3) What is a Pandas DataFrame?

Ans.) A DataFrame is a 2-dimensional, labeled data structure in Pandas that’s like a table with rows and columns (similar to Excel or SQL tables). It supports mixed data types and powerful data manipulation tools.

Q4) Explain the use of the groupby() method in Pandas

Ans.)  The groupby() method is used to split data into groups based on some criteria (like a column), apply a function (like sum, mean), and then combine the results. It’s useful for aggregation and analysis.

Q5) Why is Seaborn preferred for statistical visualizations?

Ans.) Seaborn builds on Matplotlib and makes it easier to create beautiful, informative statistical plots like boxplots, violin plots, histograms, and heatmaps with less code and better aesthetics.

Q6) What are the differences between NumPy arrays and Python lists?

Ans.) Numpy array are faster than Python lists.
      
Numpy array uses less storage than python list.

Q7) What is a heatmap, and when should it be used?

Ans.) A heatmap is a color-coded matrix used to visualize correlation, frequency, or magnitude of values. It’s ideal for showing relationships in large datasets or correlation matrices.

Q8) What does the term “vectorized operation” mean in NumPy?

Ans.) Vectorized operations mean performing operations on entire arrays without using loops. This is faster and more readable, and leverages optimized C-based implementations under the hood.

Q9) How does Matplotlib differ from Plotly?

Ans.) Matplotlib are static,low-level API whereas Plotly are interactive,high level API.

Q10) What is the significance of hierarchical indexing in Pandas?

Ans.) Hierarchical indexing allows you to have multiple index levels (MultiIndex) on rows or columns, enabling complex data structures and operations like reshaping, slicing, and aggregation on multiple axes.

Q11)What is the role of Seaborn’s pairplot() function?

Ans.) pairplot() creates a matrix of scatterplots (and histograms) for each pair of variables in a dataset. It's great for exploring relationships and distributions across multiple variables.

Q12) What is the purpose of the describe() function in Pandas?

Ans.) The describe() function provides summary statistics (count, mean, std, min, max, quartiles) for each numeric column in a DataFrame. It gives a quick overview of the dataset’s structure and distribution.

Q13.)Why is handling missing data important in Pandas?

Ans.) Missing data can lead to incorrect analysis, errors in models, or biased results. Pandas offers tools like isnull(), fillna(), and dropna() to detect and handle such cases properly.

Q14.) What are the benefits of using Plotly for data visualization?

Ans.) Interactive plots (zoom, hover, filter),Web integration (Dash, HTML export),Easy to build dashboards,Supports complex charts (3D, Geo, Heatmaps)

Q15.) How does NumPy handle multidimensional arrays?

Ans.) NumPy uses the ndarray (n-dimensional array) object which supports arrays of any dimension. You can use .reshape(), .ndim, slicing, and broadcasting to manipulate multi-dimensional data efficiently.

Q16.) What is the role of Bokeh in data visualization?

Ans.)Bokeh is a Python library for creating interactive visualizations in web browsers. It supports real-time streaming, linking plots, and building complex dashboards with ease.

Q17.) Explain the difference between apply() and map() in Pandas

Ans.) **map()**: Used on Series to apply a function element-wise.

**apply()**: Used on Series or DataFrames, can apply row-wise or column-wise functions, including lambda.

Q18.)What are some advanced features of NumPy?

Ans.) Broadcasting,Masking and filtering,Memory-efficient slicing,Linear algebra (linalg),Fourier transforms

Q19.) How does Pandas simplify time series analysis?

Ans.) Built-in support for date/time indexing,Frequency conversion (resample),Rolling window calculations,Handling time zones, periods, and offsets

Q20.)What is the role of a pivot table in Pandas?

Ans.)A pivot table lets you summarize data by reorganizing and aggregating it across two dimensions (like Excel). Useful for quickly computing grouped statistics.

Q21.) Why is NumPy’s array slicing faster than Python’s list slicing?

Ans.) NumPy arrays are stored in contiguous memory blocks and use C-level optimizations, making slicing a view (not a copy). Lists are more flexible but slower as they store references to objects.

Q22.) What are some common use cases for Seaborn?

Ans.)Visualizing correlations (heatmap),Distribution plots (distplot, kdeplot),Category comparison (boxplot, violinplot),Pairwise relationships (pairplot),Time series visualization (lineplot)

Q1.) How do you create a 2D NumPy array and calculate the sum of each row?

```python
import numpy as np

arr = np.array([[1, 2, 3], [4, 5, 6]])
row_sum = np.sum(arr, axis=1)
print("Row-wise Sum:", row_sum)
```

Q2.)Write a Pandas script to find the mean of a specific column in a DataFrame.

```python
import pandas as pd

df = pd.DataFrame({'Score': [85, 90, 75, 60]})
mean_score = df['Score'].mean()
print("Mean Score:", mean_score)
```

 Q3.) Create a scatter plot using Matplotlib.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.scatter(x, y, color='blue')
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

Q4.)  Calculate correlation matrix using Seaborn and visualize with a heatmap.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = sns.load_dataset("iris")
corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

Q5.) Generate a bar plot using Plotly.

```python
import plotly.express as px
import pandas as pd

data = pd.DataFrame({'Fruits': ['Apples', 'Bananas', 'Cherries'], 'Sales': [50, 30, 20]})
fig = px.bar(data, x='Fruits', y='Sales', title="Fruit Sales")
fig.show()
```

Q6.) Create a DataFrame and add a new column based on an existing column.

```python
import pandas as pd

df = pd.DataFrame({'Marks': [40, 60, 80, 90]})
df['Result'] = df['Marks'].apply(lambda x: 'Pass' if x >= 50 else 'Fail')
print(df)
```

Q7.) Write a program to perform element-wise multiplication of two NumPy arrays.

```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
result = a * b
print("Element-wise Multiplication:", result)
```

Q8.) Create a line plot with multiple lines using Matplotlib.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [1, 2, 3, 4]

plt.plot(x, y1, label='Squared')
plt.plot(x, y2, label='Linear')
plt.legend()
plt.title("Multiple Lines")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

Q9.) Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.

```python
import pandas as pd

df = pd.DataFrame({'Age': [15, 22, 18, 30]})
filtered_df = df[df['Age'] > 20]
print(filtered_df)
```

Q10.)  Create a histogram using Seaborn to visualize a distribution.

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("iris")
sns.histplot(data['sepal_length'], kde=True)
plt.title("Sepal Length Distribution")
plt.show()
```

Q11.) Perform matrix multiplication using NumPy.

```python
import numpy as np

a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
result = np.dot(a, b)
print("Matrix Multiplication:\n", result)
```

Q13.) Create a 3D scatter plot using Plotly.

```python
import plotly.express as px
import pandas as pd

df = px.data.iris()
fig = px.scatter_3d(df, x='sepal_length', y='sepal_width', z='petal_length',
                    color='species', title="3D Scatter Plot")
fig.show()
```
